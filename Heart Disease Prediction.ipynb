Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
# loading csv data to Pandas DataFrame
heart_data = pd.read_csv("heart_disease_data.csv")
```

```python
# printing First 5 rows
heart_data.head()
```

```python
# printing last 5 rows 
heart_data.tail()
```

```python
# Number of rows and columns in the Dataset
heart_data.shape
```

```python
# getting some informatio about the data
heart_data.info()
```

```python
# Checking for missing values
heart_data.isnull().sum()
```

```python
# Statistical Measure about the data
heart_data.describe()
```

```python
# checking the distribution of target Variable
heart_data['target'].value_counts()
```

1---> Defective Heart
0---> Healthy Heart

 Splitting Features and the target

```python
x = heart_data.drop(columns = 'target' , axis=1)
y = heart_data['target']
```

```python
print(x)
```

```python
print(y)
```

Splitting the data into training Data & testing Data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, stratify=y , random_state=2)
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

Model Training 

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training The Model with training data
model.fit(x_train,y_train)
```

Model Evaluation

```python
# Accuracy on Training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction , y_train)
```

```python
print("Accuracy on Training data : ", training_data_accuracy)
```

```python
# Accuracy on Test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction , y_test)
```

```python
print("Accuracy on Test data : ", test_data_accuracy)
```

Buildig a Predictive System

```python
input_data = (41,0,1,130,204,0,0,172,0,1.4,2,0,2)

# Change the input data to numpy array
input_data_as_numpy = np.asarray(input_data)

# reshaping The Array as we are predicting for only 1 Instance
input_data_reshaped = input_data_as_numpy.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction )

if (prediction[0] == 0):
    print("The person does not have a heart Disease")
else:
    print("The person has a heart Disease")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of target variable (Healthy vs. Defective hearts)
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=heart_data, palette='Set2')
plt.title('Distribution of Target Variable (Healthy vs. Defective Hearts)', fontsize=14)
plt.xlabel('Target (0 = Healthy Heart, 1 = Defective Heart)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks([0, 1], ['Healthy Heart', 'Defective Heart'], fontsize=10)
plt.show()
```

```python
# Pairwise correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = heart_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Pairwise Correlation Heatmap', fontsize=16)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.show()
```
